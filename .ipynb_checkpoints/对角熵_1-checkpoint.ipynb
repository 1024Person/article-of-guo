```python
import sympy as sp
from sympy.physics.quantum.spin import Jx,Jy,Jz,JzKet,JzBra
from sympy.physics.quantum import represent,Dagger,InnerProduct
from sympy.physics.quantum.qapply import qapply
from sympy import I,expand,series,Rational,print_latex,collect
import numpy as np
```

```python
t = sp.symbols('t',real=True)
N = sp.symbols('N',real=True,positive=True)
nu = sp.symbols('nu')
mu = sp.symbols('mu')
B2,B3,B8,B9,B11,B12 = sp.symbols('B_2,B_3,B_8,B_9,B_{11},B_{12}')
state_0 = JzKet(N/2,-N/2) # 初态
x11 = mu*Jx
x22 = nu*Jy
x33 = (B2*Jy+B3*Jz+B8*Jy*Jz+B9*Jz*Jy+B11*Jy**2+B12*Jz**2)
x1,x2,x3  = sp.symbols('x1:4')
e1 = sp.exp(-I*x1)
e2 = sp.exp(-I*x2)
e3 = sp.exp(I*x3)
```

```python
r1 = sp.expand(series(e1,x1,n=3)*series(e2,x2,n=3)*series(e3,x3,n=3))
# r1
```

```python
r2 = (1-I*x1-x1**2/2).subs({x1:x11})*(1-I*x2-x2**2/2).subs({x2:x22})*(1-I*x3-x3**2/2).subs({x3:x33})
# r2
```

# 验证expand函数
看看expand函数是否是按照我们想要的那种算符进行展开
$e_1*e_2$进行展开的话我们先人工展开一下试试
$$
J_x^2J_y^2+J_x^2J_y-J_x^2+J_xJ_y^2+J_xJ_y+J_x+J_y^2-J_y+1
$$

```python
expand((1-I*x1-x1**2/2).subs({x1:x11})*(1-I*x2-x2**2/2).subs({x2:x22}))
```

## 结果
很明显结果就是按照算符的展开进行计算算的，满足非对易的情况

```python
r3 = expand(r2)
# r3
```

```python
state = qapply(r3*state_0)
state
```

```python
rho = state*Dagger(state)
```

# 总结
显然到了$|\frac{N}{2},9-\frac{N}{2}\rangle$这一项的系数就是零了，所以最高就到了$|N/2,8-N/2\rangle$这一项了,同时根据限制条件$-N/2\leq 8-N/2\leq N/2$，所以
$$
-N/2\leq 8-N/2 \leq N/2
\to
\begin{cases}
    0\leq 8\\
    8\leq N
\end{cases}
$$
当然，如果$N$不满足这个条件的话，那么这一项就是零

```python
# 计算对角熵
rho00 = sp.conjugate(c0)*c0
rho11 = sp.conjugate(c1)*c1
rho22 = sp.conjugate(c2)*c2
rho33 = sp.conjugate(c3)*c3
rho44 = sp.conjugate(c4)*c4
rho55 = sp.conjugate(c5)*c5
rho66 = sp.conjugate(c6)*c6
rho77 = sp.conjugate(c7)*c7
rho88 = sp.conjugate(c8)*c8
rho = np.array([rho00,rho11,rho22,rho33,rho44,rho55,rho66,rho77,rho88])
S = sp.S(0)
for rho_item in rho:
    S+=rho_item*sp.log(rho_item) / sp.log(2)
S
```

```python
sp.factor(S)
```
