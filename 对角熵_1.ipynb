```python
import sympy as sp
from sympy.physics.quantum.spin import Jx,Jy,Jz,JzKet,JzBra
from sympy.physics.quantum import Dagger
from sympy.physics.quantum.qapply import qapply
from sympy import I,expand,series
from sympy.abc import x
import numpy as np
```

```python
t = sp.symbols('t',real=True)
N = sp.symbols('N',real=True,positive=True)
nu = sp.symbols('nu',real=True)
mu = sp.symbols('mu',real=True)
B2,B3,B8,B9,B11,B12 = sp.symbols('B_2,B_3,B_8,B_9,B_{11},B_{12}')
state0 = JzKet(N/2,-N/2) # 初态
x11 = mu*Jx
x22 = nu*Jy
x33 = B2*Jy+B3*Jz+B8*Jy*Jz+B9*Jz*Jy+B11*Jy**2+B12*Jz**2
e = sp.exp(-I*x)
ed = sp.exp(I*x)
```

```python
# r=1+I*x-x**2/2-I*x**3/6+x**4/24
# rd = 1+I*x-x**2/2-I*x**3/6+x**4/24
r=1+I*x-x**2/2
rd = 1+I*x-x**2/2
rd
```

```python
state = qapply(r.subs({x:x11})*r.subs({x:x22})*rd.subs({x:x33})*state0)
```

# 捋思路
下面捋一捋思路，现在已经将$U_3$作用在了体系的态$|N/2,-N/2\rangle$上了，下面就是整理整理每一项前面的系数

```python
c0 = qapply(JzBra(N/2,-N/2)*state)
c0
```

```python
# 开始计算对角熵
cs = [c0,c1,c2,c3,c4,c5,c6,c7,c8]
S = 0
for c in cs:
    p =c*sp.conjugate(c) 
    S+=p*sp.log(p)/sp.log(2)
```
